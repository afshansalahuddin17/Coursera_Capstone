```python
# Importing the required libraries
import torch
import numpy as np
import matplotlib.pyplot as plt
 
# A vector of size 5 with values from -3 to 3
input = np.linspace(-5, 5, 15)

output = torch.tanh(torch.FloatTensor(input))

print(input.dtype) 
print(output.dtype)
 
# Plotting
plt.plot(input, output.numpy(), color = 'blue', marker = "o")
plt.title("tanh activation function via pytorch")
plt.xlabel("X")
plt.ylabel("Y")
 
plt.show()
```

```python
# Importing the Tensorflow library
import tensorflow as tf
#enable_eager_execution
tf.config.run_functions_eagerly(True)
# A constant vector of size 8
a = tf.constant([-10.0, -7.5, -5.0, 0.0, 2.5, 5.0, 7.5, 10.0], dtype = tf.float32)
# Applying the ReLu function and
# storing the result in 'b'
b = tf.nn.relu(a, name ='ReLU')

plt.plot(a.numpy(),b.numpy(), color = 'green', marker = "o")
plt.title("ReLu activation function via pytorch")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import tensorflow as tf
from matplotlib import pyplot
pyplot.style.use('ggplot')
pyplot.figure(figsize=(10,5))
 
# rectified linear function
def Leaky_ReLU(x):
  if x>0:
	  return x
  else: 
    return 0.01*x
 
# define a series of inputs
input_series = [x for x in range(-19, 19)]
# calculate outputs for our inputs
#output_series = [Leaky_ReLU(x) for x in input_series]
output_series = [tf.nn.leaky_relu(x, alpha=0.1, name='Leaky_ReLU') for x in input_series]
# line plot of raw inputs to rectified outputs
#plt.plot(a.numpy(),b.numpy(), color = 'green', marker = "o")
pyplot.title("Leaky ReLu activation function via pytorch")
pyplot.plot(input_series, output_series,color = 'red')
pyplot.show()
```
